```python
import pathlib
import textwrap
from gtts import gTTS

from IPython.display import Audio, display
import google.generativeai as genai
import google.ai.generativelanguage as glm

# Used to securely store your API key
from google.colab import userdata

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
def ask_gemini(ques):
  modified_text = "Hey give me answer to "+ques+" in maximum of 50 words"
  response = model.generate_content(modified_text)
  to_markdown(response.text)
  return response.text
```

```python
def ask_question(name):
  ques1="hey "+name+" what do you need?"
  ques=input(ques1)
  return ques
```

```python
def classify(ques):
  web_search = False
  device_lst = ['alarm','remainder','call','messages']
  personel_lst = ['who are you?','tell me about yourself?','who created you?']

  device = False
  for i in device_lst:
    if i in ques.lower():
      device = True

  if device:
    speak("Sorry, Currently the device operations are not avaliable")
    print("Sorry, Currently the device operations are not avaliable")

  personel = False
  for i in personel_lst:
    if i in ques.lower():
      personel=True

  if personel:
    speak("I am Assistant Created by Varshitha. I can help you with your Questions")
    print("I am Assistant Created by Varshitha. I can help you with your Questions")

  if device:
    web_search=False
  elif personel:
    web_search=False
  else:
    web_search=True

  return web_search
```

```python
def speak(answer):
  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3', autoplay=True))
```

```python
have_any_questions="y"
name=''
while have_any_questions == "y":
  if name == '':
    name = input("What is your Name?")

  question = ask_question(name)
  cl = classify(question)

  answer=''
  if cl == True:
    answer = ask_gemini(question)
    speak(answer)
    print(answer)


  have_any_questions = input("Do you have any more questions?")
```
